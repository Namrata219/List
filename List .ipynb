Index and Range with list 
Slicing : Slicing a list gives us another list, instead of a single element. Slicing is an incredibly useful feature in python, one that you will use a lot!

A slice specifies a start index and an end index, and creates and returns a new list based on the indices. The indices are separated by a colon ':'.

[start: stop : step] start from the element it start, it will stop on previous element.

```python
x = list(("apple", "banana", "cherry", "mango",'tomato')) # create a list
print(x[0:5]) # Whole list, prints [1,2,3,4,5]
print(x[1:5]) #start from banana and at mango
print(x[4:5])
print(x[:]) 
```

```python
# Note that the last example[:] is often used to create duplicate lists. 
#Remember that new_list = old_list just creates a reference to old_list, but creates a copy of the old_list.
```

```python
x_new = x[:]
x_new
```

```python
x = list(("apple", "banana", "cherry", "mango",'tomato')) # indexies is always from left to right
print(x[-1:])
print(x[:-2])
print(x[-2:])
print(x[:])
print(x[::-1]) #I reverse the list order
print(x[:-1:2]) #it will count on -1 as start 
```

Using a While Loop-
You can loop through the list items by using a while loop.

Use the len() function to determine the length of the list, then start at 0 and loop your way through the list items by refering to their indexes.

Remember to increase the index by 1 after each iteration.

```python
x = ['apple', 'watermelon', 'lemon', 'orange', 'mustard']
i = 0 # this start with index 0 here apple is at 0th position.
while i < len(x):
    print(x[i])
    i = i + 1
```

Sorting the list

Sorting a list with sorted() returns a new list

sort() is a method / operation on a list which sorts the list inplace.

sort the list alphanumerically, ascending, by default:

```python
x = ['apple', 'watermelon', 'lemon', 'orange', 'mustard']
x.sort()
print(x)
```

```python
x = [22,50,11, 8, 97, 67]
x.sort()
print(x)
```

To sort descending, use the keyword argument reverse = True:

```python
x = ['apple', 'watermelon', 'lemon', 'orange', 'mustard']
x.sort(reverse= True)
print(x)
```

```python
x = [22,50,11, 8, 97, 67]
x.sort(reverse= True) #list.sort()
print(x)
```

```python
x = [22,50,11, 8, 97, 67]
x1=sorted(x) #sorted() we have to assign it to new value. 
print(x1)
```

One important thing to note is that sort() modifies the initial variable directly, 
and consequently, the initial order will be lost.

On the other hand, sorted() keeps a copy of the initial variable, 
making it possible to revert to the initial order if needed. 
Because sort() does not make any copy of the initial variable, it is a bit more efficient than sorted().
However, this comes at the cost of convenience.

```python
##Customize Sort Function
##You can also customize your own function by using the keyword argument key = function.

##The function will return a number that will be used to sort the list (the lowest number first):
```

```python
def item(n):
    return abs(n)
x= [22,50,11,8,97,67]
x.sort(key =item)
print(x)
```

```python
def item(n):
    return abs(n-50) #this will minus the item from x with 50 and will be arrange in acsending order. 
x= [22,50,11,8,97,67]
x.sort(key =item)
print(x)
```

To remove the duplicates from a list, you can make use of the built-in function set(). 
The specialty of the set() method is that it returns distinct elements.

You can remove duplicates from the given list by importing OrderedDictfrom collections.

It is available from python2.7 onwards. OrderedDictdict takes care of returning you the 
distinct elements in an order in which the key is present.

You can make use of a for-loop that we will traverse the list of items to remove duplicates.

The method unique() from Numpy module can help us remove duplicate from the list given.

The Pandas module has a unique() method that will give us the unique elements from the list given.

The combination of list comprehension and enumerate is used to remove the duplicate elements from the list. 
Enumerate returns an object with a counter to each element in the list.

```python
#Set method
x= [1,1,2,3,4,5,'55',21,12,12,9,'namu']
print("Before removal of duplicate ", x)
x1=set(x)
print("After removal" ,list( x1))
```

```python
#Temporary list
x= [1,1,2,3,4,5,'55',21,12,12,9,'namu']
print("List before removing duplicate"  ,x)
x1=[]

for i in x:
    if i not in x1:
        x1.append(i)
x=x1
print("After removal" ,x1)
```
